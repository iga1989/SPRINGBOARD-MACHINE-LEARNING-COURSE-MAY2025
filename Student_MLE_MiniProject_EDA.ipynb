```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load parquet file into a Pandas DataFrame
df = pd.read_parquet('yellow_tripdata_2022-01.parquet')
```

```python
# Display the first few rows of the dataset
df.head()
```

```python
# Display the shape of the dataset
df.shape
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
df.describe()
```

```python
# Drop rows with missing values.
df_no_nulls = df.dropna().copy()
```

```python
df_no_nulls.shape
```

```python
#Check for the data type of columns
df_no_nulls.dtypes
```

```python
# Create a new column for trip duration in minutes
df_no_nulls.loc[:, 'trip_duration'] = (
    (df_no_nulls['tpep_dropoff_datetime'] - df_no_nulls['tpep_pickup_datetime']).dt.total_seconds()/60
).astype(int)
```

```python
# Preview the result tpep_dropoff_datetime tpep_pickup_datetime  tpep_trip_duration_in_minutes
print(df_no_nulls[['tpep_dropoff_datetime', 'tpep_pickup_datetime', 'trip_duration']].head())
```

```python
# Create new columns for pickup hour and day of week
# Extract pickup hour (0-23)
df_no_nulls.loc[:, 'pickup_hour'] = df_no_nulls['tpep_pickup_datetime'].dt.hour
# Extract day of week (Monday=0, Sunday=6)
df_no_nulls.loc[:, 'dayofweek'] = df_no_nulls['tpep_pickup_datetime'].dt.dayofweek
```

```python
df_no_nulls.head()
```

```python
print(df_no_nulls[['tpep_pickup_datetime', 'pickup_hour', 'dayofweek', 'trip_duration']].head())
```

```python
df_no_nulls.dtypes
```

```python
# Create a lineplot displaying the number of trips by pickup hour
#Group by pickup hour and count trips
trip_counts = df_no_nulls.groupby('pickup_hour').size().reset_index(name='trip_count')
#Plotting
plt.figure(figsize=(10,6))
sns.lineplot(data=trip_counts, x='pickup_hour', y='trip_count', marker='o')
plt.title('Number of Trips by Pickup Hour')
plt.xlabel('Pickup Hour')
plt.ylabel('Number of Trips')
plt.xticks(range(0,24))  # ensures all hours are shown
plt.grid(True)

#Show the plot
plt.show()
```

```python
# Create a lineplot displaying the number of trips by pickup day
#Group by pickup hour and count trips
trip_counts = df_no_nulls.groupby('dayofweek').size().reset_index(name='trip_count')
#Plotting
plt.figure(figsize=(10,6))
sns.lineplot(data=trip_counts, x='dayofweek', y='trip_count', marker='o')
plt.title('Number of Trips by Pickup day')
plt.xlabel('Pickup Day')
plt.ylabel('Number of Trips')
plt.xticks(range(0,7))  # ensures all days are shown
plt.grid(True)

#Show the plot
plt.show()
```

```python
# Compute correlation matrix of numerical variables
cols = ['trip_distance', 'fare_amount', 'tip_amount', 'total_amount', 'trip_duration']
df_corr = df_no_nulls[cols]
corr_matrix = df_corr.corr()
print(corr_matrix)
```

```python
# Create a heatmap of the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Selected Variables')
plt.show()
# No variable are strongly related according to the heatmap
```

```python
# Create a scatter plot matrix of numerical variables. If memory issues try the df.sample method.
numerical_df = df_no_nulls.select_dtypes(include=['float64', 'int64']).sample(frac=0.1, random_state=1)
sns.pairplot(numerical_df)
plt.show()
```

```python
# Create a Seaborn countplot for PULocationID and DOLocationID. Only plot the top 15 categories by value counts.
```

```python
# Get top 15 most frequent pick-up locations
top_pu = df_no_nulls['PULocationID'].value_counts().nlargest(15).index
print(top_pu)
# Filter for top 15 pick-up locations
df_top_pu = df_no_nulls[df_no_nulls['PULocationID'].isin(top_pu)]
print(df_top_pu)
```

```python
# Get top 15 most frequent drop-off locations
top_do = df_no_nulls['DOLocationID'].value_counts().nlargest(15).index
print(top_do)
# Filter for top 15 drop-off locations
df_top_do = df_no_nulls[df_no_nulls['DOLocationID'].isin(top_do)]
print(df_top_do)
```

```python
# Plot pick-up location countplot
plt.figure(figsize=(12, 6))
sns.countplot(x='PULocationID', data=df_top_pu, order=top_pu)
plt.title('Top 15 Pick-up Locations')
plt.show()
```

```python
# Plot drop-off location countplot
plt.figure(figsize=(12, 6))
sns.countplot(x='DOLocationID', data=df_top_do, order=top_do)
plt.title('Top 15 Drop-off Locations')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='payment_type', y='total_amount', data=df_no_nulls)
plt.ylim(0, 1000)  # Limit the y-axis to $1,000
plt.title('Total Amount by Payment Type')
plt.xlabel('Payment Type')
plt.ylabel('Total Amount ($)')
plt.show()
```

```python
# Choose data source: sample_df if sampled, or df_no_nulls otherwise
sample_df = df_no_nulls.sample(frac=0.1, random_state=42)  # Take 10% sample

plt.figure(figsize=(15, 4))

# Plot fare_amount
plt.subplot(1, 3, 1)
sns.histplot(data=sample_df, x='fare_amount', bins=50, kde=True)
plt.title('Fare Amount Distribution')

# Plot trip_distance
plt.subplot(1, 3, 2)
sns.histplot(data=sample_df, x='trip_distance', bins=50, kde=True)
plt.title('Trip Distance Distribution')

# Plot extra
plt.subplot(1, 3, 3)
sns.histplot(data=sample_df, x='extra', bins=50, kde=True)
plt.title('Extra Charges Distribution')

plt.tight_layout()
plt.show()
```
